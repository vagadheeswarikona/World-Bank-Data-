#Merge the all tables(Life_expectancy, Fertility_rate, Country_population, Metadata_Country)


```python
df_Metadata_Country.head()
```

```python
df_Metadata_Country = df_Metadata_Country[['Country Code', 'Region']]
df_Metadata_Country.head()
```

```python
df_Metadata_Country.isna().sum()
```

```python
country_region_mapping = df_Metadata_Country.dropna().set_index('Country Code')['Region'].to_dict()
df_Metadata_Country['Region'] = df_Metadata_Country.apply(
    lambda row: country_region_mapping.get(row['Country Code'], row['Region']) if pd.isna(row['Region']) else row['Region'], axis=1
)
```

```python
df_Metadata_Country.duplicated().sum()
```

```python
df_Metadata_Country.dropna(inplace=True)
df_Metadata_Country.shape
```

```python
df_merge1 = pd.merge(df_Metadata_Country,df_Country_Population_Melt,how='left', on=['Country Code'])
df_merge1.head()
```

```python
df_merge1.shape
```

```python
df_merge2= pd.merge(df_merge1,df_Life_Excetancy_Melt,how='left', on=['Country Code','Year','Country Name'])
df_merge2.head()
```

```python
df_merge3= pd.merge(df_merge2,df_Fertility_rate_Melt,how='left', on=['Country Code','Year','Country Name'])
df_merge3.head()
```

```python
df_merge3.shape
```

```python
df_merge3.isna().sum()
```

```python
df_merge3.duplicated().sum()
```

```python
df_merge3.describe()
```

```python
df_merge3.dtypes
```

```python
for index, row in df_merge3.iterrows():
    if pd.isna(row['Population']):
        previous_value = df_merge3.loc[index - 1, 'Population'] if index > 0 else np.nan
        next_value = df_merge3.loc[index + 1, 'Population'] if index < len(df_merge3) - 1 else np.nan
        if not pd.isna(previous_value) and not pd.isna(next_value):
            df_merge3.at[index, 'Population'] = (previous_value + next_value) / 2
```

```python
df_merge3.isna().sum()
```

```python
import matplotlib.pyplot as plt
df_merge3['Year'] = pd.to_numeric(df_merge3['Year'])
population_trends = df_merge3.groupby('Year')['Population'].sum()
# Splitting the data into four parts
part1 = population_trends.loc['1960':'1974']
part2 = population_trends.loc['1975':'1989']
part3 = population_trends.loc['1990':'2004']
part4 = population_trends.loc['2005':'2016']

# Plotting the line graphs for each part
plt.figure(figsize=(20, 12))

plt.subplot(2, 2, 1)
part1.plot(kind='line')
plt.title('Population Trends (1960-1974)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

plt.subplot(2, 2, 2)
part2.plot(kind='line')
plt.title('Population Trends (1975-1989)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

plt.subplot(2, 2, 3)
part3.plot(kind='line')
plt.title('Population Trends (1990-2004)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

plt.subplot(2, 2, 4)
part4.plot(kind='line')
plt.title('Population Trends (2005-2016)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
for index, row in df_merge3.iterrows():
    if pd.isna(row['Life_Expectancy']):
        # Check for previous and next values
        previous_value = df_merge3.loc[index - 1, 'Life_Expectancy'] if index > 0 else np.nan
        next_value = df_merge3.loc[index + 1, 'Life_Expectancy'] if index < len(df_merge3) - 1 else np.nan

        # Fill if both neighbors are not NaN
        if not pd.isna(previous_value) and not pd.isna(next_value):
            df_merge3.at[index, 'Life_Expectancy'] = (previous_value + next_value) / 2
        # If only one neighbor is available
        elif not pd.isna(previous_value):
            df_merge3.at[index, 'Life_Expectancy'] = previous_value
        elif not pd.isna(next_value):
            df_merge3.at[index, 'Life_Expectancy'] = next_value
```

```python
import seaborn as sns

# Plotting a distribution plot for life expectancy
plt.figure(figsize=(10, 6))
sns.histplot(df_merge3['Life_Expectancy'].dropna(), kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.show()
```

```python
for index, row in df_merge3.iterrows():
    if pd.isna(row['Fertility_Rate']):
        # Check for previous and next values
        previous_value = df_merge3.loc[index - 1, 'Fertility_Rate'] if index > 0 else np.nan
        next_value = df_merge3.loc[index + 1, 'Fertility_Rate'] if index < len(df_merge3) - 1 else np.nan

        # Fill if both neighbors are not NaN
        if not pd.isna(previous_value) and not pd.isna(next_value):
            df_merge3.at[index, 'Fertility_Rate'] = (previous_value + next_value) / 2
        # If only one neighbor is available
        elif not pd.isna(previous_value):
            df_merge3.at[index, 'Fertility_Rate'] = previous_value
        elif not pd.isna(next_value):
            df_merge3.at[index, 'Fertility_Rate'] = next_value
```
